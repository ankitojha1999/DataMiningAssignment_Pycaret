```python
import pycaret
pycaret.__version__
```

```python
from pycaret.datasets import get_data
data = get_data('gold')
data = data.iloc[:200]
data.columns
data[['Crude Palm Oil_T-1']].plot()
```

```python
print(', '.join(data.columns))
```

```python
from pycaret.time_series import *
s = setup(data, target='Silver_T-22', fh = 10, session_id = 123)  # Forecasting horizon of 10 days
```

```python
from pycaret.time_series import TSForecastingExperiment
exp = TSForecastingExperiment()
```

```python
type(exp)
```

```python
exp.setup(data, target='Silver_T-22', fh = 10, session_id = 123)
```

```python
check_stats()
```

```python
best = compare_models()
```

```python
exp.compare_models()
```

```python
plot_model(best, plot = 'forecast')
```

```python
pred_obj = predict_model(best)
```

```python
pred_obj.head()
```

```python
save_model(best, 'GoldPipeline')
```
