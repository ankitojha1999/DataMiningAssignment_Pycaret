```python
# check installed version (must be >3.0)
import pycaret
pycaret.__version__
```

##Pycaret **Dataset**

```python
from pycaret.datasets import get_data
data = get_data('index')
```

```python
### load sample dataset from pycaret dataset module
from pycaret.datasets import get_data
data = get_data('parkinsons')
```

##Pycaret Regression Model

```python
# import pycaret regression and init setup
from pycaret.regression import *
s = setup(data, target = 'test_time', session_id = 123)
```

# Model Comparisons

```python
best = compare_models()
```

```python
best = create_model('et') #choose catboost to avoid plot errors with some models
```

## Plot Analysis

```python
# plot residuals
plot_model(best, plot = 'residuals')
```

```python
best = create_model('lightgbm')
```

```python
# plot residuals
plot_model(best, plot = 'residuals')
```

```python
# plot error
plot_model(best, plot = 'error')
```

```python
# plot feature importance
plot_model(best, plot = 'feature')
```

```python
evaluate_model(best)
```

```python
# predict on test set
holdout_pred = predict_model(best)
```

```python
# show predictions df
holdout_pred.head()
```

```python
# copy data and drop test_time

new_data = data.copy()
new_data.drop('test_time', axis=1, inplace=True)
new_data.head()
```

```python
# predict model on new_data
predictions = predict_model(best, data = new_data)
predictions.head()
```
